```python
#Q.2
df1=pd.DataFrame({
    'Id':[1,2,3],
    'Name':['Harsh','Harshita','Kunal']
})
df2=pd.DataFrame({
    'Id':[2,3,4],
    'Branch': ['CS','DS','IOT']
})

#inner merge
inner=df1.merge(df2, on='Id',how='inner')
print ("\nInner Join:-\n",inner)

#left join
left=df1.merge(df2, on='Id',how='left')
print("\nLeft Join:-\n",left)

#right join
right=df1.merge(df2, on='Id',how='right')
print("\nRight Join:-\n",right)

#index based joining
df1_index=df1.set_index('Id')
df2_index=df2.set_index('Id')
index_join=df1_index.join(df2_index)
print("\nIndex based joining:-\n",index_join)

#Merge on multiple keys

df1['Semester']=[3,3,3]
df2['Semester']=[3,4,3]
mul_merge=df1.merge(df2,how='inner', on =['Id','Semester'])
print("\nMerge on multiple keys:\n",mul_merge)
```

```python
#Q3:
df1 = pd.DataFrame({
    'ID': [1, 2],
    'Name': ['Harsh', 'Harshita']
})

df2 = pd.DataFrame({
    'ID': [3, 4],
    'Name': ['Kunal', 'Krishna']
})

df3 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'City': ['Delhi', 'Mumbai', 'Chennai', 'Kolkata']
})

#vertical concatinate df1 and df2
df_concat=pd.concat([df1,df2],axis=0,ignore_index=True)
print("\nVertical concatinate df1 and df2:-\n",df_concat)

#Merge with 3rd df
result=df_concat.merge(df3,on='ID')
print ("\nMerging 2df:-\n",result)

#Joining with 3rd df
df_concat=df_concat.set_index('ID')
df3=df3.set_index('ID')
result=df_concat.join(df3)
print("\nJoining 2 df:-\n",result)

#Major difference is merge works on coloumns and join works on index
```

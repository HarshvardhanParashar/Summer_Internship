```python
#Q.1
name= input("Enter your name: ")
classs= input("Enter your class: ")

print("Enter marks of 5 subjects: ")
sst=int(input("Social Science: "))
eng=int(input("English: "))
sci=int(input("Science: "))
math=int(input("Math: "))
comp=int(input("Computer: "))

total= sst+eng+sci+math+comp
percentage= (total/500)*100
print("Result of the student is: ")
print("Name: ",name)
print("Class: ",classs)
print("Total Marks: ",total)
print("Percentage: ",percentage)
```

```python
#Q.2
str1= input("Enter first string: ")
str2= input("Enter second string: ")
str3= str1+str2
print("Concatenated string: ",str3)
print("Length of concatenated string: ",len(str3))
print("Lowercase: ",str3.lower())
print("Uppercase: ",str3.upper())
print("Title Case: ",str3.title())
print("Swapcase: ",str3.swapcase())
print("Capitalize: ",str3.capitalize()) 
print("Casefold: ",str3.casefold())
print("Centered: ",str3.center(50))
print("Count of 'a': ",str3.count('a'))
print("Ends with 'a': ",str3.endswith('a'))
print("Find 'a': ",str3.find('a'))
print("Isalnum: ",str3.isalnum())
print("Isdigit: ",str3.isdigit())
print("Isnumeric: ",str3.isnumeric())
print("Isspace: ",str3.isspace())
print("Replace 'a' with '@': ",str3.replace('a','@'))
```

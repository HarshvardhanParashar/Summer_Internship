```python
#Q.1
#Csv file for address book 2 3 dummy data

import csv
with open('address_book.csv','a',newline='') as file:
    writer= csv.writer(file)
    while True:
        name= input("Enter name: ")
        phone= input("Enter phone number: ")
        email= input("Enter email: ")
        writer.writerow([name,phone,email])
        break
```

```python
#Q.2
#Create database
import sqlite3
conn= sqlite3.connect('students.db')

#Table creation
conn.execute('''CREATE TABLE IF NOT EXISTS students
                (id INTEGER PRIMARY KEY AUTOINCREMENT,
                name TEXT NOT NULL);''')
print("Table created successfully.\n")

conn.execute('''CREATE TABLE IF NOT EXISTS marks
                (id INTEGER PRIMARY KEY AUTOINCREMENT,
                student_id INTEGER,
                subject TEXT NOT NULL);''')
print("Table 2 created successfully.\n")

conn.execute('''Create table if not exists m_no
             (id integer primary key autoincrement,
             m_no integer not null);''')
print("Table 3 created successfully.\n")
```

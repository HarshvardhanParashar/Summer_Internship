```python
import pandas as pd
#Q1
#Series from dictionary
data={"day1":420,"day2":350}
qmarks=pd.Series(data)
print(qmarks)
print(type(qmarks))
#Series from list
dt=[1,2,3]
s_dt=pd.Series(dt)
print(s_dt)
#Access elements of series
print(s_dt[1])
print(qmarks["day2"])
```

```python
#Q.2

data = [
    [1, "Harsh", 85],
    [2, "Aman", 90],
    [3, "Riya", 78]
]

df = pd.DataFrame(data)

print(df)

#DataFrame from Python Dictionary
data = {
    "ID": [1, 2, 3],
    "Name": ["Harsh", "Aman", "Riya"],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

print(df)

# DataFrame Using List of Lists
data = [
    [101, "Laptop", 55000],
    [102, "Mouse", 700],
    [103, "Keyboard", 1200]
]

df = pd.DataFrame(data, columns=["Product_ID", "Product_Name", "Price"])

print(df)

# DataFrame Using List of Tuples
data = [
    (1, "Delhi", 200),
    (2, "Mumbai", 300),
    (3, "Jaipur", 150)
]

df = pd.DataFrame(data, columns=["ID", "City", "Sales"])

print(df)

# DataFrame from List of Dictionaries

data = [
    {"Name": "Harsh", "Age": 20, "City": "Beawar"},
    {"Name": "Aman", "Age": 21, "City": "Jaipur"},
    {"Name": "Riya", "Age": 19, "City": "Kota"}
]

df = pd.DataFrame(data)

print(df)
```

```python
#Q.3
import pandas as pd
data = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Harsh", "Aman", "Riya"],
    "Marks": [85, 90, 78]
})

#iterate rows
for i in range(len(data)):
    print(data.iloc[i]["Name"])

#Result based on condition
res=data[data['Marks']>80]
print(res)

#Selecting rows based on iloc
res2= data.loc[0:1,['Name','Marks']]
print (res2)

#Drop rows
res3= data.drop(data[data['Marks']>80].index)
print(res3)

#insert at specific posn
new_row = pd.DataFrame({
    "Name": ["Neha"],
    "Age": [23],
    "Marks": [92]
})

df1 = data.iloc[:2]
df2 = data.iloc[2:]

data = pd.concat([df1, new_row, df2]).reset_index(drop=True)

print(data)

#list from rows
row_list = data.values.tolist()

print(row_list)
```

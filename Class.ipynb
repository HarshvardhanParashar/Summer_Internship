```python
#importing and connecting
import sqlite3
conne=sqlite3.connect("example.db")

#creating an table
table_query='''
    CREATE TABLE emp2(
        id integer primary key autoincrement,
        name varchar(50),
        dept varchar(50)
    )'''
conne.execute(table_query)
conne.commit()
#insert data 
insert_query='''
    Insert into emp (id,name,dept) values
    (1,"kunal","data science"),
    (2,"harsh","mechanical")
'''
conne.execute(insert_query)
conne.commit()
conne.close()
```

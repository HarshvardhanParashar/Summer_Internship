```python
#Q.1
def find_maximum(num1,num2,num3):
    if num1>=num2 and num1>=num3:
        return num1
    elif num2>=num1 and num2>=num3:
        return num2
    else:
        return num3
    
num1= int(input("Enter first number: "))
num2= int(input("Enter second number: "))
num3= int(input("ENter third number: "))
print("Maximum number is:", find_maximum(num1,num2,num3))
```

```python
#Q.2
def new_list(list):
    unique_list= []
    for i in list:
        if i not in unique_list:
            unique_list.append(i)
    return unique_list

list=[]
n= int(input("Enter number of elements in the list: "))
for i in range(n):
    element= input("Enter element: ")
    list.append(element)
print("Original List: ",list)
print("List with unique elements: ",new_list(list))
```

```python
#Q.3
def mul_list(list):
    result=1
    for i in list:
        result*=i
    return result
list=[]
n= int(input("Enter number of elements in the list: "))
for i in range(n):
    element= int(input("Enter element: "))
    list.append(element)
print("Original List: ",list)
print("Product of all elements in the list: ",mul_list(list))
```

```python
#Q.4
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
number= int(input("Enter a number to find its factorial: "))
print("Factorial of",number,"is",factorial(number))
```

```python
#Q.6
def  range_check(num,lower,upper):
    if num>=lower and num<=upper:
        print(f"{num} is within the range {lower} to {upper}.")
    else:
        print(f"{num} is outside the range {lower} to {upper}.")
num= int(input("Enter a number: "))
lower= int(input("Enter lower bound: "))
upper= int(input("Enter upper bound: "))
range_check(num,lower,upper)
```

```python
#Q.7
def even_numbers(list):
    even_list= []
    for i in list:
        if i%2==0:
            even_list.append(i)
    return even_list    
list=[]
n= int(input("Enter number of elements in the list: "))
for i in range(n):
    element= int(input("Enter element: "))
    list.append(element)
print("Original List: ",list)
print("Even numbers in the list: ",even_numbers(list))
```

```python
#Q.8
def is_prime(num):
    if num<=1:
        return False
    for i in range(2,int(num**0.5)+1):
        if num%i==0:
            return False
    return True
number= int(input("Enter a number to check if it is prime: "))
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

```python
#Q.9
#Count upper and lower case letters in a string
def count_case(string):
    upper_count= 0
    lower_count= 0
    for char in string:
        if char.isupper():
            upper_count+=1
        elif char.islower():
            lower_count+=1
    print(f"Uppercase letters: {upper_count}")
    print(f"Lowercase letters: {lower_count}")
string= input("Enter a string: ")
count_case(string)
```

```python
#Q.10
#Reading, writing and appending to a file
filename= input("Enter filename: ")

#Writing to a file
with open(filename,'w') as file:
    content= input("Enter content to write to the file: ")
    file.write(content)
#Appending to a file
with open(filename,'a') as file:
    content= input("Enter content to append to the file: ")
    file.write("\n"+content)

#Reading from a file
with open(filename,'r') as file:
    print("Content of the file: ")
    print(file.read())
```

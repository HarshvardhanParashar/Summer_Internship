```python
name= input("Enter your name: ")
classs= input("Enter your class: ")

print("Enter marks of 5 subjects: ")
sst=int(input("Social Science: "))
eng=int(input("English: "))
sci=int(input("Science: "))
math=int(input("Math: "))
comp=int(input("Computer: "))

total= sst+eng+sci+math+comp
percentage= (total/500)*100
```

```python
#Q.1
#Practice Dictionary, Tuples and Set
#Dictionary
student= {
    "name": name,
    "class": classs,
    "marks": {
        "Social Science": sst,
        "English": eng,
        "Science": sci,
        "Math": math,
        "Computer": comp
    },
    "total": total,
    "percentage": percentage
}
print("Student Dictionary: ",student)
print("Student Name: ",student["name"])
print("Student Class: ",student["class"])
print("Marks in Social Science: ",student["marks"]["Social Science"])
print("Total Marks: ",student["total"])

#Tuples
subjects= ("Social Science", "English", "Science", "Math", "Computer")
print("Subjects Tuple: ",subjects)
print("First Subject: ",subjects[0])
print("Last Subject: ",subjects[-1])
print("Number of Subjects: ",len(subjects))

#Sets
unique_subjects= set(subjects)
print("Unique Subjects Set: ",unique_subjects)
```

```python
#Q.3
def palindrome_check(string):
    if string == string[::-1]:
        print("The string is a palindrome.")
    else:
        print("The string is not a palindrome.")
    
input_string= input("Enter a string to check for palindrome: ")
palindrome_check(input_string)
```

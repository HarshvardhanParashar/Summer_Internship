```python
#spark session
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("SalesDataFrameAnalysis") \
    .master("local[*]") \
    .getOrCreate()

print("Spark Version:", spark.version)
```

```python
df = spark.read.csv(
    "/workspace/sales_data.csv",
    header=True,
    inferSchema=True
)

df.show()
```

```python
#sort by sales
sorted_df = df.orderBy(
    df.sales.desc()
)

sorted_df.show()
```

```python
#top 3 highest sales
top3 = sorted_df.limit(3)

top3.show()
```

```python
#sales >80000
high_sales_df = df.filter(
    df.sales > 80000
)

high_sales_df.show()
```

```python
#Saving file
high_sales_df.write \
    .mode("overwrite") \
    .option("header", True) \
    .csv("/workspace/output/high_sales_products.csv")

print("CSV saved successfully")
```
